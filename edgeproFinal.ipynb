```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
import requests
import pandas as pd
```

```python
def get_link(choice,state):
    state = state.replace(" ",'%20')
    template = 'https://www.edgeprop.my/{}/{}/all-commercial?page='
    link = template.format(choice,state)
    return link
```

```python
def main(choice,state):
        link = get_link(choice,state)
        driver = webdriver.Chrome(executable_path='chromedriver.exe')
        driver.maximize_window()
        records=[]
        temp_list=[]
        scraped_count = 0
        for page in range(1,3):
                newlink=link+str(page)
                driver.get(newlink)
                
                time.sleep(3)

                # records=[]
                # temp_list=[]
                # scraped_count = 0
                print(newlink)

                all_cards = driver.find_elements(By.CLASS_NAME,'card.tep-listing-card')
                print(len(all_cards))
                
                for card in all_cards:
                        all_cards = driver.find_elements(By.CLASS_NAME,'card.tep-listing-card')
                        #print(all_cards)
                        # card = all_cards[i]

                        link = card.find_element(By.TAG_NAME, 'a')
                        clink = link.get_attribute('href')
                        #print(clink)

                        try:
                                title = card.find_element(By.CLASS_NAME, "listing-name").text
                        except Exception as e:
                                title = "no title"
                        #print(title)

                        try:
                                market_value = card.find_element(By.CLASS_NAME, "listing-price").text
                        except Exception as e:
                                market_value = "no market value"
                        #print(market_value)

                        try:
                                area = card.find_element(By.CLASS_NAME, "listing-address").text
                        except Exception as e:
                                area = "no area"
                        #print(area)
                        
                        try:
                                img = card.find_element(By.TAG_NAME, 'img')
                                pImg = img.get_attribute('src')
                        except Exception as e:
                                pImg = "no image"
                        #print(pImg)

                        temp_list.append((clink, title, market_value, area, pImg))
                        

                for a in temp_list:
                        #print('a ',a)

                        driver.get(a[0]); 
                        time.sleep(3) 

                        try:
                                details = driver.find_element(By.CLASS_NAME,'detail-info-sight.row').text.strip() 
                        except Exception as e:
                                details = "no detail"
                                #print(e)
                        #print(details)

                        try:
                                description = driver.find_element(By.CLASS_NAME,'d-none.d-md-block').text.strip() 
                        except Exception as e:
                                description = "no description"
                                #print(e)
                        #print(description)
                        
                        records.append(
                                (a[0],a[1],a[2],a[3],a[4],details,description)
                        )
                        scraped_count+=1
                        print("scraped_count",scraped_count)
                
                link = get_link(choice,state)
        driver.quit()                
            
        col = [ 'Link','Titile','Market_value','Area','Image', 'details', 'description']
        commercial_data = pd.DataFrame(records ,columns=col)
        commercial_data.to_csv('Commercial_edge.csv')        
```

```python
choice = input('Enter buy/rent: ')
state = input('Enter state you are looking for : ')
main(choice,state) 
 
```
